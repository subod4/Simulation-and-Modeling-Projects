```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define Predator-Prey Equations (Lotka-Volterra Model)
def predator_prey(y, t, alpha, beta, delta, gamma):
    prey, predator = y
    dydt = [alpha * prey - beta * prey * predator,  # Prey growth
            delta * prey * predator - gamma * predator]  # Predator growth
    return dydt

# Parameters
alpha = 0.1  # Prey birth rate
beta = 0.02  # Predation rate
delta = 0.01  # Predator reproduction rate
gamma = 0.1  # Predator death rate

# Initial Conditions
prey0 = 40   # Initial prey population
predator0 = 9   # Initial predator population
y0 = [prey0, predator0]

# Time Simulation (in arbitrary units)
t = np.linspace(0, 200, 1000)  # Simulate for 200 time units

# Solve the Differential Equations
solution = odeint(predator_prey, y0, t, args=(alpha, beta, delta, gamma))
prey_pop, predator_pop = solution.T

# Plot Results
plt.figure(figsize=(10, 5))
plt.plot(t, prey_pop, label="Prey Population", color="blue")
plt.plot(t, predator_pop, label="Predator Population", color="red")
plt.xlabel("Time")
plt.ylabel("Population Size")
plt.title("Predator-Prey Model (Lotka-Volterra Equations)")
plt.legend()
plt.grid()
plt.show()
```
